# Resnet 50 summit

```python
import pandas as pd
import numpy as np
import pickle
```

```python
import os, sys
# Open a file
path = "/scratch/paulcalle/kidney_results/summit/archRESNET50_results"
dirs = os.listdir( path )
```

```python
dirs
```

```python
import re

l_time_total_files = []

r = re.compile('time_total')

for item in dirs:

    res = r.search(item)
    if res != None:
        l_time_total_files.append(item)
```

```python
l_time_total_files
```

```python
df_time_total_resnet50 = pd.DataFrame(np.nan, index=["K%s_outer"%i for i in np.arange(1,11,1)],
                     columns=["k%s_val"%i for i in np.arange(1,11,1)])

max_epoch = 0 

for item in l_time_total_files:
    print(item)
    # Get outter look
    r_outer = re.compile(r'(?<=K)[0-9]+')
    k_outer = r_outer.search(item).group(0)
    print("k_outer = ", k_outer)

    # Get inner look
    r_inner = re.compile(r'(?<=k)[0-9]+')
    k_val = r_inner.search(item).group(0)
    print("k_val = ", k_val)
    
    
    total_temp = np.load('/scratch/paulcalle/kidney_results/summit/archRESNET50_results/' + item)
    
    print(total_temp)    
    # Search index of validation
          
    df_time_total_resnet50.loc["K%s_outer"%k_outer,"k%s_val"%k_val] = total_temp
    
print("max epoch = ",max_epoch)
```

```python
df_time_total_resnet50
```

```python
df_time_total_resnet50.to_csv("resnet50_time.csv")
```
